# **Data Analysis**

now we beginning the analysis of the cleaned data. what we going to do here are doing analytical statistics and the calculation to boost the argument from the **Problem Identification**

```python
# for statistics process to fill section of Data Analysis
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import skew, kurtosis
```

the cell code above is to utilise the graphics and diagrams for showing the visualisation of the analytical statistics

- **Rating**

if the client wants to buy the smartphone based on rating that matches with his desire, we can see the analytical statistics and the visualisation below.

```python
# Working Area

# Rating Column
print("Rating\n")
print(f"Mean: {df_1['Rating'].mean():.2f}")
print(f"Median: {df_1['Rating'].median():.2f}")
print(f"Mode: {df_1['Rating'].mode().values[0]:.2f}")
print(f"Standard Deviation: {df_1['Rating'].std():.2f}")
print(f"Skewness: {df_1['Rating'].skew()}")
print(f"Kurtosis: {df_1['Rating'].kurtosis()}")
```

- The 'Rating' data has a mean of 4.23, indicating an overall positive sentiment.
- The median and mode of the 'Rating' data are both 4.30, suggesting a concentration of ratings around this value.
- The standard deviation of 0.27 indicates that the ratings are relatively close to the mean, reflecting consistency in the ratings.
- The skewness of -1.33 suggests that the distribution of ratings is negatively skewed, meaning it's slightly skewed towards higher values.
- The kurtosis of 4.11 indicates that the distribution of ratings has heavy tails and is leptokurtic, meaning it has more extreme values compared to a normal distribution.

```python
# Create a figure side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Plot skewness
sns.histplot(df_1['Rating'], kde=True, ax=axes[0])
axes[0].set_title('Skewness of Rating')
axes[0].set_xlabel('Rating')
axes[0].set_ylabel('Frequency')

# Plot kurtosis
sns.boxplot(df_1['Rating'], ax=axes[1])
axes[1].set_title('Kurtosis of Rating')
axes[1].set_xlabel('Rating')
axes[1].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

the graphics above shows the overall Rating from that matches with his desire and his consideration to buy a smartphone without classified into each brands. the next graphics is to show the classification of each brands for the rating to get more information.  

```python
# Group the data by 'Brands'
grouped = df_1.groupby('Brands')

# Count the number of ratings per brand
rating_counts = grouped.size().sort_values(ascending=False)

# Select the top 10 brands
top_10_brands = rating_counts.head(10).index

# Filter the dataframe to include only the top 10 brands
df_top_10 = df_1[df_1['Brands'].isin(top_10_brands)]

# Group the filtered dataframe by 'Brands'
grouped_top_10 = df_top_10.groupby('Brands')

# Calculate skewness and kurtosis for 'Rating' within each group
skewness = grouped_top_10['Rating'].apply(skew)
kurt = grouped_top_10['Rating'].apply(kurtosis)

# Create a figure side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Plot skewness
sns.barplot(x=skewness.index, y=skewness.values, ax=axes[0])
axes[0].set_title('Skewness of Rating by Top 10 Brands')
axes[0].set_xlabel('Brands')
axes[0].set_ylabel('Skewness')
axes[0].tick_params(axis='x', rotation=45)

# Plot kurtosis
sns.barplot(x=kurt.index, y=kurt.values, ax=axes[1])
axes[1].set_title('Kurtosis of Rating by Top 10 Brands')
axes[1].set_xlabel('Brands')
axes[1].set_ylabel('Kurtosis')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

the graphics above is quite messed up, so we will use another way to get the information-- it's about ***Top 10 Branded Smartphones based on Ratings*** here are the visualisation graphics below: 

```python
# Select the top 10 smartphone brands based on ratings
top_10_brands = df_1.groupby('Brands')['Rating'].mean().nlargest(10)

# Filter the data for the top 10 brands
df_top_10 = df_1[df_1['Brands'].isin(top_10_brands.index)]

# Create a visualization for the top 10 smartphone brands based on ratings
plt.figure(figsize=(10, 6))
sns.boxplot(x='Brands', y='Rating', data=df_top_10, order=top_10_brands.index)
plt.title('Top 10 Smartphone Brands Based on Ratings')
plt.xlabel('Brand')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()
```

the box plot above shows both information-- the good one and the bad one. the good one is the safe zone which is some mobile based on the brand has popular rating so the client secured to decide which smartphone is the best for him before dealing with the price. the outliers represent the least ratings due to some minor testimonials, bad reviews, and some negative things for the smartphone's overall review from the market. 

```python
# Calculate skewness for 'Rating' within each smartphone brand
skewness = df_top_10.groupby('Brands')['Rating'].skew()

# Create a bar plot to visualize skewness
plt.figure(figsize=(10, 6))
sns.barplot(x=skewness.index, y=skewness.values)
plt.title('Skewness of Ratings for Top 10 Smartphone Brands')
plt.xlabel('Brand')
plt.ylabel('Skewness')
plt.xticks(rotation=45)
plt.show()
```

# **Analytical Statistics Inferential**

in this section we want to see the possibility for the client to fulfill his demand which is to get a smartphone based on his desire.  

```python
from scipy.stats import ttest_ind

# Define low budget and high budget groups
low_budget_df = df_1[df_1['Selling Price'] <= 100000]
high_budget_df = df_1[df_1['Selling Price'] > 100000]

# Extract selling prices for both groups
low_budget_prices = low_budget_df['Selling Price']
high_budget_prices = high_budget_df['Selling Price']

# Perform two-sample independent t-test
t_statistic, p_value = ttest_ind(low_budget_prices, high_budget_prices)

# Determine the significance
alpha = 0.05
if p_value < alpha:
    print("Client can buy a smartphone he desires.")
else:
    print("Client cannot buy a smartphone he desires.")
```

basically the codes above indicates two sample independent test from the information we have about smartphones and its informations. the client has budget equals to 100000 Rupees which means he can buy a smartphone he desires among all provided smartphones that matches with his needs, and align with his budget.
